DETECTING THE OUTLIERS

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(data[['bright_ti4','scan','track','bright_ti5','frp',]])
```

```python
print(data.shape)
```

```python
# bright_ti4
Q1 = np.percentile(data['bright_ti4'], 25)
Q2= np.percentile(data['bright_ti4'], 50)
Q3 = np.percentile(data['bright_ti4'], 75)
IQR = Q3 - Q1
lower_bti4 = Q1 - 1.5 * IQR
upper_bti4 = Q3 + 1.5 * IQR
bright_ti4_outliers = data[(data['bright_ti4'] < lower_bti4) | (data['bright_ti4'] > upper_bti4)]
print("Outliers in bright_ti4:")
print(bright_ti4_outliers[['bright_ti4']])

# frp
Q1_frp = np.percentile(data['frp'], 25)
Q2= np.percentile(data['frp'], 50)
Q3_frp = np.percentile(data['frp'], 75)
IQR_frp = Q3_frp - Q1_frp
lower_frp = Q1_frp - 1.5 * IQR_frp
upper_frp = Q3_frp + 1.5 * IQR_frp
frp_outliers = data[(data['frp'] < lower_frp) | (data['frp'] > upper_frp)]
print("Outliers in frp:")
print(frp_outliers[['frp']])

# scan
Q1_scan = np.percentile(data['scan'], 25)
Q2= np.percentile(data['scan'], 50)
Q3_scan = np.percentile(data['scan'], 75)
IQR_scan = Q3_scan - Q1_scan
lower_scan = Q1_scan - 1.5 * IQR_scan
upper_scan = Q3_scan + 1.5 * IQR_scan
scan_outliers = data[(data['scan'] < lower_scan) | (data['scan'] > upper_scan)]
print("Outliers in scan:")
print(scan_outliers[['scan']])
```

REMOVING THE OUTLIERS

```python
data = data[
    (data['bright_ti4'] >= lower_bti4) & (data['bright_ti4'] <= upper_bti4) &
    (data['bright_ti5'] >= np.percentile(data['bright_ti5'], 25) - 1.5 * (np.percentile(data['bright_ti5'], 75) - np.percentile(data['bright_ti5'], 25))) &
    (data['bright_ti5'] <= np.percentile(data['bright_ti5'], 75) + 1.5 * (np.percentile(data['bright_ti5'], 75) - np.percentile(data['bright_ti5'], 25))) &
    (data['frp'] >= lower_frp) & (data['frp'] <= upper_frp)
]
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import folium
import gradio as gr
import requests
```

```python

def show_fire_on_map_html(latitude, longitude):
    fire_map = folium.Map(location=[latitude, longitude], zoom_start=10)
    folium.Marker(
        location=[latitude, longitude],
        popup=" Fire Detected!",
        icon=folium.Icon(color='red', icon='fire')
    ).add_to(fire_map)
    return fire_map._repr_html_()

def find_fire_stations_text(lat, lon, radius=40000):
    query = f"""
    [out:json];
    node(around:{radius},{lat},{lon})["amenity"="fire_station"];
    out;
    """
    stations_text = ""
    try:
        response = requests.post("https://overpass-api.de/api/interpreter", data={"data": query})
        response.raise_for_status()
        data = response.json()

        elements = data.get("elements", [])
        if not elements:
            stations_text = "ðŸš« No fire stations found nearby."
        else:
            for element in elements:
                name = element["tags"].get("name", "Unnamed Fire Station")
                latitude = element["lat"]
                longitude = element["lon"]
                stations_text += f"ðŸš’ {name}\n"
                stations_text += f"ðŸ“ Latitude: {latitude}, Longitude: {longitude}\n"
                stations_text += "----------\n"

    except requests.exceptions.RequestException as e:
        stations_text = f"âš ï¸ Error fetching data from Overpass API: {e}"

    return stations_text

def predict_fire_gradio(latitude, longitude, bright_ti4, scan, track, bright_ti5, frp):
    input_data = pd.DataFrame([{
        'latitude': latitude,
        'longitude': longitude,
        'bright_ti4': bright_ti4,
        'scan': scan,
        'track': track,
        'bright_ti5': bright_ti5,
        'frp': frp
    }])

    # Scale input values
    scale_cols = ['scan', 'track', 'bright_ti4', 'bright_ti5', 'frp']
    input_data[scale_cols] = scaler.transform(input_data[scale_cols])

    # Make prediction
    prediction = best_rf_model.predict(input_data)

    prediction_text = " Prediction Result:\n"
    map_html = None
    fire_stations_info = ""

    if prediction[0] == 1:
        prediction_text += " Fire Detected!"
        map_html = show_fire_on_map_html(latitude, longitude)
        fire_stations_info = find_fire_stations_text(latitude, longitude)
    else:
        prediction_text += " No Fire Detected."

    geolocator = Nominatim(user_agent="simple_fire_app")
    try:
        location = geolocator.reverse((latitude, longitude), language='en', timeout=10)
        if location:
            prediction_text += f"\n Location: {location.address}"
        else:
            prediction_text += "\n Location: Not found"
    except requests.exceptions.ReadTimeout:
        prediction_text += "\n Location: Geocoding service timed out. Could not retrieve location."
    except Exception as e:
        prediction_text += f"\n An error occurred during geocoding: {e}"

    return prediction_text, map_html, fire_stations_info

# Gradio Interface
inputs = [
    gr.Number(label="Latitude"),
    gr.Number(label="Longitude"),
    gr.Number(label="Bright_ti4"),
    gr.Number(label="Scan"),
    gr.Number(label="Track"),
    gr.Number(label="Bright_ti5"),
    gr.Number(label="FRP")
]

outputs = [
    gr.Textbox(label="Prediction and Location Info"),
    gr.HTML(label="Fire Location Map"),
    gr.Textbox(label="Nearby Fire Stations")
]

gr.Interface(
    fn=predict_fire_gradio,
    inputs=inputs,
    outputs=outputs,
    title="Forest Fire Prediction System",
    description="Enter the required parameters to predict if there is a forest fire."
).launch(debug=True)
```
